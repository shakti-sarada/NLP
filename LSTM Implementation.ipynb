```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
# droppong nan values
df =df.dropna()
```

```python
## Get the Independent Features

X=df.drop('label',axis=1)
```

```python
## Get the Dependent features
y=df['label']
```

```python
y.shape
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
### Vocabulary size
voc_size=5000
```

# Onehot Representation

```python
messages=X.copy()
```

```python
messages['title'][1]
```

```python
messages
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
messages.reset_index(inplace=True) # reset the DataFrame's index to a default range
```

```python
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer ##stemming purpose
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus] 
onehot_repr
```

# Embedding Representation

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='post',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[1]
```

```python
## Creating model
embedding_vector_features=40 ##features representation
model=Sequential()
# adding embedding layer
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
# adding lstm layer
model.add(LSTM(100))
# adding dense layer
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
len(embedded_docs),y.shape
```

```python
import numpy as np
X_final=np.array(embedded_docs) # converting embedded_docs from list to array
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

# Model Training

```python
# Training of data
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

# Performance Metrics And Accuracy

```python
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred > 0.5, 1,0) ##use AUC ROC Curve to find the probability or threshold for y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

# Adding Dropout 

```python
# dropout is used to reduce overfitting
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```
